```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno
from datetime import datetime, date
```

```
mobile = pd.read_csv('train.csv')
```

```
#Drop duplicates if any
mobile.drop_duplicates(inplace=True)
```

```
# compute the correlation matrix
corr = mobile.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(10, 5))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=0.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}).set(title='Correlation Matrix')
plt.show()
```

```
#px width/height
ax = sns.regplot(x="px_width", y="px_height", data=mobile)

ax.set_xlabel("Pixel Resolution Width",fontsize=20)
ax.set_ylabel("Pixel Resolution Heigh",fontsize=20)
sns.set(rc={'figure.figsize':(30,20)})
```

```
#sch scw
ax = sns.regplot(x="sc_h", y="sc_w", data=mobile)

ax.set_xlabel("Screen Height",fontsize=40)
ax.set_ylabel("Screen Width",fontsize=40)
sns.set(rc={'figure.figsize':(10,8)})
```

```
#pc fc
ax = sns.regplot(x="pc", y="fc", data=mobile)

ax.set_xlabel("Primary Camera",fontsize=20)
ax.set_ylabel("Front Camera",fontsize=20)
sns.set(rc={'figure.figsize':(10,8)})
```

```
#3g 4g
ax = sns.regplot(x="three_g", y="four_g", data=mobile)
ax.set_xlabel("3G",fontsize=20)
ax.set_ylabel("4G",fontsize=20)
sns.set(rc={'figure.figsize':(10,8)})
```

```
#price pxheight
sns.catplot(x="price_range", y="px_height", kind="point", data=mobile).set(title='The Relationship between Price and PX Height')
sns.set(rc={'figure.figsize':(15,8)})
```

```
#price pxwidth
sns.catplot(x="price_range", y="px_width", kind="point", data=mobile).set(title='The Relationship between Price and PX Width')
sns.set(rc={'figure.figsize':(15,8)})
```

```
#battery power and price
sns.catplot(x="price_range", y="battery_power", kind="bar", data=mobile).set(title='The Relationship between Price and Battery Power')
sns.set(rc={'figure.figsize':(15,8)})
```
